```python
from konlpy.tag import Okt
from konlpy.tag import Postprocessor

def tokenize_with_konlpy_okt(text):
    okt = Okt()
    tokens = okt.morphs(text)
    return tokens

# 엑셀 파일 불러오기
file_path = 'faang_review.xlsx'
df = pd.read_excel(file_path)

# Columns to tokenize
columns_to_tokenize = ['페이스북', '아마존', '애플', '넷플릭스', '구글']

# Dictionary to store tokenized data for each column
tokenized_data = {}

# Perform tokenization for each specified column
for column in columns_to_tokenize:
    column_tokens = df[column].apply(lambda x: tokenize_with_konlpy_okt(x))
    tokenized_data[column] = column_tokens

# Print tokenized data for each column
for column, tokens in tokenized_data.items():
    print(f"Konlpy Okt tokens for {column}:", tokens)

ngrams = [(('네', '임밸류'), 'Noun'), (('업', '무량'), 'Noun'), (('워', '라', '밸'), 'Noun'), (('워', '라', '벨'), 'Noun'), (('비', '추천'), 'Noun')]

# Postprocessor를 사용하여 ngram을 하나의 토큰으로 결합
postprocessor = Postprocessor(ngrams)
for column, tokens in tokenized_data.items():
    tokenized_data[column] = postprocessor.decompound(tokens)

# 수정된 토큰화된 결과물 출력
for column, tokens in tokenized_data.items():
    print(f"Modified Konlpy Okt tokens for {column}:", tokens)
```

```python
import pandas as pd
import re

# 불용어 처리 + 지정어 처리
def clean_text(text, stopwords):
    for stopword in stopwords:
        text = text.replace(stopword, '')
    return text

file_path = 'faang_review.xlsx'
df = pd.read_excel(file_path)

columns = ['페이스북', '아마존', '애플', '넷플릭스', '구글']
stopwords = {'내가', '한', '되어', '돠어', '와', '에', '으로', '에서', '하고', '을', '를', '위한', '+', ':', '/', '(', ')', '?', ';', '!'}

# Stopwords 추가
for column in columns:
    stopwords.add(column)

# Stopwords를 컬럼별로 적용
for column in columns:
    df[column] = df[column].apply(lambda x: clean_text(str(x), stopwords))

replace = {
    '워라벨': '워라밸',
    '워밸' : '워라밸'
    '워' : '워라밸',
    '라' : '워라밸,'
    '벨류': '밸류',
    '컬처': '문화',
    '컬쳐': '문화',
    '나이스': '좋은',
    '팀바팀': '팀바이팀',
    '부바부': '팀바이팀',
}

# Replacements를 컬럼별로 적용
for column in columns:
    df[column] = df[column].replace(replace, regex=True)

# 정제된 데이터프레임을 새로운 CSV 파일로 저장
df.to_csv('cleaned_faang_review.csv', index=False)

# CSV 파일을 다시 데이터프레임으로 불러오기 (확인용)
cleaning_df = pd.read_csv('cleaned_faang_review.csv')

print(cleaning_df.head(5))
```

```python
#네이버 단점 + 내향
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

# 성격 요인 문서 예시 (임의로 만든 것입니다)
personality_documents = [
    "조직경계 개방적",
    "의사결정 다수 참여",
    "결정하다 의견일치",
    "회의 많다",
    " 부서 협업 많다",
    "외부전문가 의존한"
]

# 기업 리뷰 문서 예시 (임의로 만든 것입니다)
company_review_documents = [
   "매우 수직적 탑다운 문화",
   "유저 고민 손 빠르면 됨",
   "어떠한 전력 기획 청사진 없다",
   "히스토리 관리 엉망 파악 어려우 묻는 알아서 파악 원함"
]

# 문서에 태그 부여 (성격 요인 문서에는 "personality", 기업 리뷰 문서에는 "company_review")
tagged_personality_documents = [TaggedDocument(words=document.split(), tags=["personality"]) for document in personality_documents]
tagged_company_review_documents = [TaggedDocument(words=document.split(), tags=["company_review"]) for document in company_review_documents]

# 모델 학습
model = Doc2Vec(vector_size=50, window=2, min_count=1, workers=4, epochs=100)
tagged_documents = tagged_personality_documents + tagged_company_review_documents
model.build_vocab(tagged_documents)
model.train(tagged_documents, total_examples=model.corpus_count, epochs=model.epochs)

# 성격 요인 문서와 기업 리뷰 문서 간의 유사도 비교
similarity_score = model.dv.similarity("personality", "company_review")
print(f"Similarity Score: {similarity_score}")
```
