```python
#넷플릭스 토큰화 및 csv
import pandas as pd
from konlpy.tag import Okt

def tokenize_with_konlpy_okt(text):
    okt = Okt()
    tokens = okt.pos(text)  # Get part-of-speech tagging
    filtered_tokens = [word for word, pos in tokens if pos in ['Noun', 'Adjective']]
    combined_tokens = ' '.join(filtered_tokens)  # Join filtered tokens into a single string
    return combined_tokens

# 엑셀 파일 불러오기
file_path = 'C:/Users/a5752/Downloads/faang_review.xls'
df = pd.read_excel(file_path)

# Columns to tokenize
columns_to_tokenize = ['넷플릭스']

# Dictionary to store tokenized data for each column
tokenized_data = {}

# Perform tokenization for each specified column
for column in columns_to_tokenize:
    column_tokens = df[column].apply(lambda x: tokenize_with_konlpy_okt(x))
    tokenized_data[column] = column_tokens

# Convert the tokenized data dictionary into a DataFrame
tokenized_df = pd.DataFrame(tokenized_data)

# Save the DataFrame to a CSV file
csv_output_path = 'Netflix_tokenized_data.csv'
tokenized_df.to_csv(csv_output_path, index=False, encoding='utf-8-sig')  # utf-8-sig for proper handling of Korean characters

# Print the DataFrame
print("Tokenized Data DataFrame:")
print(tokenized_df)

# Print a message about the CSV file
print(f"\nTokenized data saved to: {csv_output_path}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from konlpy.tag import Okt
import re

# 나눔 폰트 경로 설정
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'

```

```python
# 기업의 인재상 텍스트 데이터
corporate_values = df.iloc[:, 1]

# 불용어 처리 함수
def text_preprocessor(corporate_values, stop_words=None):
    # (1) [], (), {}, <> 괄호와 괄호 안 문자 제거하기
    pattern = r'\([^)]*\)'  # ()
    corporate_values = re.sub(pattern=pattern, repl='', string=corporate_values)

    pattern = r'\[[^)]*\]'  # []
    corporate_values = re.sub(pattern=pattern, repl='', string=corporate_values)

    pattern = r'\<[^)]*\>'  # <>
    corporate_values = re.sub(pattern=pattern, repl='', string=corporate_values)

    pattern = r'\{[^)]*\}'  # {}
    corporate_values = re.sub(pattern=pattern, repl='', string=corporate_values)

    # (2) '...외', '...총' 제거하기
    corporate_values = corporate_values.replace('', ' ')

    # (3) 특수문자 제거
    pattern = r'[^a-zA-Z가-힣]'
    corporate_values = re.sub(pattern=pattern, repl=' ', string=corporate_values)

    # (4) 단위 제거: cm, km, etc.
    units = ['mm', 'cm', 'km', 'ml', 'kg', 'g']
    for unit in units:
        corporate_values = corporate_values.replace(unit, '')

    # (5) 형태소 분석
    okt = Okt()
    s_morphs = okt.morphs(corporate_values)

    # (6) 글자 1개만 있으면 제외하기
    s_list = [word for word in s_morphs if len(word) != 1]

    # (7) 불용어 처리
    if stop_words:
        s_list = [word for word in s_list if word not in stop_words]

    return s_list

# 불용어 예시: '아무', '가보지', '않은' 등을 불용어로 추가할 수 있습니다.
stop_words = []

# 텍스트 전처리
processed_text = ' '.join(text_preprocessor(corporate_values.iloc[0], stop_words)).lower()

# 토큰화 및 품사 태깅
tokenized_sentence = word_tokenize(processed_text)
tagged_words = pos_tag(tokenized_sentence)

# 명사, 동사, 형용사 추출
selected_words = [word for word, pos in tagged_words if pos in ['NN', 'VB', 'JJ']]

# 빈도 분석
word_count = pd.Series(selected_words).value_counts()

# TF-IDF 분석
tfidf = TfidfVectorizer()
tdm = tfidf.fit_transform([processed_text])
tfidf_values = pd.DataFrame(tdm.toarray(), columns=tfidf.get_feature_names_out())

# 워드클라우드 생성
wordcloud = WordCloud(font_path=font_path, background_color='white', width=800, height=400).generate_from_frequencies(word_count)
plt.figure(figsize=(10, 5))

# 시각화
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
